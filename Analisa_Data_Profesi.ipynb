```python
### jika ingin melihat occupation yang dibedakan berdasarkan gender
dist = pd.crosstab(df.occupation,df.gender)
dist
```

```python
## menambahkan total antara F dan M (gender)

dist['total'] = dist['F']+dist['M']
dist
```

```python
## Membuat persentase dari gender tersebut

dist['% Female'] = dist['F']/dist['total'] *100
dist['% Male'] = dist['M']/dist['total'] *100
dist['% Total'] = dist['total']/dist['total']*100
```
